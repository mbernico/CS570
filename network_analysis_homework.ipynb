```python
import networkx as nx
%matplotlib inline
```

```python
M=nx.read_gml("polblogs.gml", relabel=True)
```

```python
# create weighted graph from M
G = nx.Graph()
for u,v,data in M.edges_iter(data=True):
    if G.has_edge(u,v):
        G[u][v]['weight'] += 1
    else:
        G.add_edge(u, v, weight=1)

```
